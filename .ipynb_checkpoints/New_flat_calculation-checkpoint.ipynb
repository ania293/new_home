## Buying new home

#### Prices of a house
Based on the market research the current price of our dream home is about 120 000zl. We are looking to buy a new flat in 5 years.
How the price of our flat will change within the next five years knowing that the real estate value in our neighbourhood will keep increasing by 5% annually? 
Assumptions:
prices will be updated monthly

```python
## Load libraries
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
present_value = 120000
years = 5
n_period = 5*12
annual_growth_rate = 0.05
monthly_growth_rate = (1+annual_growth_rate)**(1/12)-1
periods = np.arange(1,n_period+1)

# Calculate future value over 5 years period
fv_home_overtime = -npf.fv(monthly_growth_rate, periods, 0, present_value)

# Assumed price of a house in 5 years
fv_home_5y = -npf.fv(monthly_growth_rate, n_period, 0, present_value)
fv_home_5y
print(f"Predicted price of the flat in 5 years is {round(fv_home_5y,2)}")
```

#### Investment plan
In order to buy a home in 5 years, we have to create a plan to accumulate enough money. Luckily, we have found an amazing deposit with an annual interest rate of 12% compounded monthly. 
How much money has to be saved each month to afford our dream home in the next five years?

```python
current_save = 0
annual_intrest_rate = 0.12
monthly_interest_rate = (1+annual_intrest_rate)**(1/12)-1

required_monthly_saving = -npf.pmt(monthly_interest_rate, n_period, 0, fv_home_5y)

print(f"We will have to save {round(required_monthly_saving,2)}zl each month to be able to buy a home in 5 years.")

#investment_plan = np.zeros(n_period)
#for i in range(n_period):
#    if i == 0:
#        prev_investment = 0
#    else:
#        prev_investment = investment_plan[i-1]    
#    investment_growth = prev_investment * (1+ monthly_interest_rate)
#    investment_plan[i] = required_monthly_saving + investment_growth


savings_plan = -npf.fv(monthly_interest_rate, periods, required_monthly_saving, 0)
```

#### Visualisation
The line plot presents prices of the flat in our neighbourhood and the investment growth over the period of 5 years.


```python
plt.plot(fv_home_overtime,label='Flat value')
plt.plot(savings_plan,label='Investment/Savings growth')
plt.legend()
plt.xlabel('Number of periods')
plt.ylabel('Money in zloty')
```
